```python
from google.cloud import bigquery
import pandas as pd
```

```python
# Especifique o caminho para o arquivo JSON de credenciais
credencial = r"C:\Users\107\Downloads\teste-big-query-123456-282b8bc713a3.json"
# Crie um cliente BigQuery usando as credenciais
client = bigquery.Client.from_service_account_json(credencial)
```

```python
# QUERYS
querySADT_Digit_Manual = """ 
SELECT * 
  FROM `teste-big-query-123456.SADT.SADT_Digit_Manual_012024`
 LIMIT 25
"""
querySADT_Digit_Portal=""" 
SELECT *
  FROM `teste-big-query-123456.SADT.SADT_Digit_Portal_012024` 
 WHERE CD_PRESTADOR_EXECUTOR = 256
 LIMIT 10
"""
querySADT_ImportA500="""
SELECT * 
  FROM `teste-big-query-123456.SADT.SADT_ImportA500_012024`
  LIMIT 15
"""
querySADT_ImportXML="""
SELECT * 
  FROM `teste-big-query-123456.SADT.SADT_ImportXML_012024`
  LIMIT 20
"""
```

```python
# Criando variavel com resultado da Query
resultadoSADT_Digit_Manual = client.query(querySADT_Digit_Manual)
resultadoSADT_Digit_Portal = client.query(querySADT_Digit_Portal)
resultadoSADT_ImportA500 = client.query(querySADT_ImportA500)
resultadoOSADT_ImportXML = client.query(querySADT_ImportXML)
```

```python
# Converta os resultados em um DataFrame
SADT_Digit_Manual = resultadoSADT_Digit_Manual .to_dataframe()
SADT_Digit_Portal = resultadoSADT_Digit_Portal.to_dataframe()
SADT_ImportA500 = resultadoSADT_ImportA500 .to_dataframe()
SADT_ImportXML = resultadoOSADT_ImportXML.to_dataframe()
```

```python
# Definir o caminho e nome dos arquivos
pasta = r"\\10.8\usuarios$\107\Meus Documentos\Extracao_BigQuery\\"
Excel_SADT_Digit_Manual = 'SADT_Digit_Manual.xlsx'
Excel_SADT_Digit_Portal = 'SADT_Digit_Portal.xlsx'
Excel_SADT_ImportA500 = 'SADT_ImportA500.xlsx'
Excel_SADT_ImportXML = 'SADT_ImportXML.xlsx'


# Função para remover informações de fuso horário das colunas datetime
def remove_timezone(df):
    for col in df.select_dtypes(['datetimetz']):
        df[col] = df[col].dt.tz_localize(None)
    return df

# Remover informações de fuso horário de cada DataFrame
SADT_Digit_Manual = remove_timezone(SADT_Digit_Manual)
SADT_Digit_Portal = remove_timezone(SADT_Digit_Portal)
SADT_ImportA500 = remove_timezone(SADT_ImportA500)
SADT_ImportXML = remove_timezone(SADT_ImportXML)

# Escrever nos arquivos Excel
SADT_Digit_Manual.to_excel(pasta + Excel_SADT_Digit_Manual, index=False)
SADT_Digit_Portal.to_excel(pasta + Excel_SADT_Digit_Portal, index=False)
SADT_ImportA500.to_excel(pasta + Excel_SADT_ImportA500, index=False)
SADT_ImportXML.to_excel(pasta + Excel_SADT_ImportXML, index=False)
```
